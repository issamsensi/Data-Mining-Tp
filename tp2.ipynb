```python
# l'importation des bibliotheques
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

hs = pd.read_csv('Housing.csv')
```

```python
# distribution des prix des maisons
sns.histplot(hs["price"], bins=30)
plt.show()
```

on remarque d'apres l'histogramme que c'est une distribusion asymtrie a droite

```python
# distribution de la superficie
sns.histplot(hs["area"], bins=30)
plt.show()
```

et voila c'est la meme chose pour la superficie aussi, c'est une distribution asymetrique a droite

```python
# distribution des bathrooms et bedrooms
fig, ax = plt.subplots(1, 2)

ax[0].hist(hs["bathrooms"])
ax[0].set_title("distribution des bathrooms")

ax[1].hist(hs["bedrooms"])
ax[1].set_title("distribution des bedrooms")

plt.show()
```

on remarque la meme chose pour ces duex, une distribution asymetrie a droite

```python
# la relation entre la superficie (area) et le prix (price)

fig, ax = plt.subplots(1, 2, figsize=(14,5))

sns.scatterplot(x="area", y="price", data=hs, ax=ax[0])
ax[0].set_title("La relation entre la superficie (area) et le prix (price)")
ax[0].set_xlabel("Area (superficie)")
ax[0].set_ylabel("Price")

sns.regplot(x="area", y="price", data=hs, ax=ax[1])
ax[1].set_title("Relation en utilisant la regression linéaire")
ax[1].set_xlabel("Area (superficie)")
ax[1].set_ylabel("Price")

plt.tight_layout()
plt.show()


corr = hs["area"].corr(hs["price"])
print("Correlation entre la surface et le prix:", corr)
```

d'apres le scaterplot et la regression liniere et aussi la valeur de corelation qui est 0.5359973457780797 on remarque qu'on a une corelation positive moyenne entre ces des facteurs, c'est a dire si la surface augmente le prix augment aussi wa l3akso sahih si le prix augmente fa hada yobayin une augmentation sur la surface aussi.

```python
# la relation entre les bedrooms et le prix 

fig, ax = plt.subplots(1, 2, figsize=(14,5))

sns.scatterplot(x="bedrooms", y="price", data=hs, ax=ax[0])
ax[0].set_title("La relation entre les bedrooms et le prix")
ax[0].set_xlabel("bedrooms")
ax[0].set_ylabel("Price")

sns.regplot(x="bedrooms", y="price", data=hs, ax=ax[1])
ax[1].set_title("Relation en utilisant la regression linéaire")
ax[1].set_xlabel("bedrooms")
ax[1].set_ylabel("Price")

plt.tight_layout()
plt.show()


corr = hs["bedrooms"].corr(hs["price"])
print("Correlation entre les bedrooms et le prix:", corr)
```

On remarque une corelation positive faible cette fois

```python
# la relation entre la climatisation (airconditioning) et le prix (price)

sns.boxplot(x="airconditioning", y="price", data=hs)
plt.title("Relation entre Airconditioning et Price")
plt.show()
```

On remarque d'apres le boxplot que les maisons qui on un climatiseurs a plus de prix (plus cheres)

```python
# l'impact de l'ameublement (furnishingstatus) sur le prix (price)

sns.boxplot(x="furnishingstatus", y="price", data=hs)
plt.title("Relation entre furnishingstatus et Price")
plt.show()
```

on remarque que les maisons furnished sont plus cheres que les autres, et les sumi-furnished plus que unfurnished, alors le prix est dans cette order: furnished > semi-furnished > unfurnished

```python
# la matrice de corelation
corr = hs.select_dtypes(include=['int64', 'float64']).corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matrice de corelation")
plt.show()
```

d'apres la matrice de corelation les variables plus corelles avec le prix sont: area (surface) et bathrooms

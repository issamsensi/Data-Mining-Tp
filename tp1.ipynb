```python
import pandas as pd
import numpy as np

#import data and show the structure

hs = pd.read_csv('Housing.csv')

print(hs.head())

print('\n\n')

hs.info()
```

**after this we'll explore the dataset**

```python
#nombre des colonnes et des lignes
print(f'number of lines: {hs.shape[0]}\nnumber of columns: {hs.shape[1]}')

#les colonnes presentes dans le dataset
print('\nColumns in the dataset:')
for col in hs.columns:
    print(col, end=', ')

print('\n')
#les types des donnes
print(hs.dtypes)

#colonnes numeriques et categorique
numerical_cols = hs.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = hs.select_dtypes(include=['object']).columns
print('\nNumerical columns:', end=' ')
for col in numerical_cols:
    print(col, end=', ')
print('\nCategorical columns:', end=' ')
for col in categorical_cols:
    print(col, end=', ')
```

**Informations et Statistique Descriptive du Dataset**

```python
# les valeurs manquantes
print(f'\n les valeurs manquantes:\n{hs.isnull().sum()}')
```

on remarque que notre data et 'clean', pas des valeurs manquantes

```python
# les indicateurs statistiques globaux des colonnes numériques
print(hs.select_dtypes(include=['int64', 'float64']).describe())
```

```python
# premier et troisieme quartile
print('la premiere quartile des valeurs numeriques: ')
print(hs.quantile(q=0.25, axis=0, numeric_only=True, interpolation='linear'))
print('\nla troixieme quartile des valeurs numeriques: ')
print(hs.quantile(q=0.75, axis=0, numeric_only=True, interpolation='linear'))

```

```python
# la moyenne et le median des prix
print(f'la moyenne des prix: {hs['price'].mean()}')
print(f'\nle median des prix: {hs['price'].median()}')
```

```python
# la répartition des catégories pour chaque variable catégorielle

# print(hs.select_dtypes(include=['object']).value_counts())
# print(hs.select_dtypes(include=['object']).describe())


for col in hs.select_dtypes(include="object"):
    print(hs[col].value_counts())
    #hs[col].value_counts().plot(kind="bar")
    print()
```

```python
# Combien de valeurs manquantes y a-t-il dans chaque colonne
# we see that there is no valeurs manquantes d'apres les questions precedants

print(f'\n les valeurs manquantes:\n{hs.isnull().sum()}')
```

```python
# les 5 quartiers avec les prix de vente moyens les plus élevés
print('les 5 quartiers avec les prix de vente moyens les plus élevés: \n')
print(hs.sort_values(by='price', ascending=False).head())
```

```python
# le prix moyen par statut d'ameublement
print(f'le prix moyen par statut d\'ameublement: \n')
print(hs.groupby('furnishingstatus')['price'].mean())
```

```python
'''
Quelle est la moyenne des prix pour les maisons ayant : 
La climatisation (airconditioning = yes) ? 
Un sous-sol (basement = yes) ? 
Un accès à la route principale (mainroad = yes) ?
'''
print('this moyenne: ', end='')
print(hs[(hs['basement'] == 'yes') & 
   (hs['airconditioning'] == 'yes') & 
   (hs['mainroad'] == 'yes')]['price'].mean())
```

```python
# la moyenne des prix en fonction du nombre d'étages 
print('la moyenne des prix en fonction du nombre d\'étages: ')
print(hs.groupby('stories')['price'].mean())
```
